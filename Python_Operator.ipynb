# Logic Operator

```python
# a and b
# a or b
# not 

print("True and False: ",True and False)
print("True or False: ",True or false )
print("not True", not True)
```

```python
a = -10
b = 20
if a>0 and b>0:
    print("a+b =", a+b)
else:
    print("a-b =", a-b)
```

```python
a = -50
b = -25
if a>0 or b>0:
    print("a//b = ", a//b)
else:
    print("a*b = ", a*b)
```

```python
print(10 or 20)
print(30 or 40)

print(10 and 20)
print(30 and 40)

print(not 10)
print(not 0)
```

# In and Not in Operator

```python
numbers = [12,10,2,6,8]
print(6 in numbers)
print(32 in numbers)
print(51 not in numbers)
print(12 not in numbers)
```
