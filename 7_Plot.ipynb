# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(170,10,250)
# print(x)
plt.hist(x)
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([25,35,46,36])
mylabel = ["A","B","C","D"]
myexplode=[0.2,0,0,0]
mycolors = ["black","hotpink","green","red"]
# plt.pie(y, labels=mylabel, startangle=90)
plt.pie(y,labels=mylabel,explode= myexplode,
          startangle=90,
          shadow=True, colors=mycolors)
plt.legend(title="Alephba")
plt.show()
```

# Bar Chart

```python
x = ["A","B","C","D"]
y = [3,10,15,2]
z=[2,5,6,3]
plt.bar(x,y)
plt.bar(x,z)
plt.show()
```
