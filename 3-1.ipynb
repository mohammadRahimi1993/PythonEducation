# Example_2: Square of numbers

```python
numbers = [1,2,3,4,5,6]
sum = 0
for item in numbers:
    #print(item**2, end=" ")
    sum += item**2
print(sum)
```

# Example_3:Avrage of numbers

```python
numbers = [1,2,3,4,5,6]
sum = 0
for item in numbers:
    sum += item
list_size = len(numbers)
avg = sum/list_size
print(avg)
```

# Example_4:Disply odd and even numbers

```python
numbers = list(range(1,200))
for item in numbers:
    if item % 2 == 0:
        # print(f"{item} is even")
        print("item is even.")
    else:
        print(f"{item} is odd")
```

# Example_5: Sum of even numbers

```python
numbers = list(range(1,50))
sum = 0
for item in numbers:
    if item % 2 == 0:
        sum += item
print(sum)
```
