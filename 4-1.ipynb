```python
# Example_6
sentence = "Hello python"
count = 0
for char in sentence:
    if char != "l":
        continue
    else:
        count = count + 1
print("Total of l is: ",count)
        
```

# Else block in for loop

```python
# Example_8
for num in range(1,6):
    print(num, end=" ")
else:
    print("Done")
    
# Example_9
for num in range(1,20):
    if num < 15:
        print(num, end=" ")
    else:
        continue
else:
    print("FOR loop was run.")
```

# Reverse for loop

```python
# Example_10
for num in range(6,1,-1):
    print(num, end=" ")
print(" ")

# Example_11
my_list = [1,2,3,"s","A","Python"]
for item in reversed(my_list):
    print(item, end=" ")
print(" ")

# Example_12
for item in my_list[::-1]:
    print(item, end=" ")
```

# Slice in loop

```python
numbers = [1,2,3,4,5,6]
print(numbers[2])
print(numbers[1:4])
print(numbers[1:4:1])
print(numbers[1:4:2])
print(numbers[4:1:-1])
```

# Iterate String using for loop

```python
# Example_1
sentence = "Hello python"
for char in sentence:
    print(char, end=" ")
print(" ")    

# Example_2
for i, v in enumerate(sentence):
    if(i >= 6 and i <= 10):
        print(v,end="")
print(" ")

# Example_3
for char in sentence[::-1]:
    print(char,end="")
print(" ")  

# Exampe_4
sentence = "I'm student."
words = sentence.split()
print(words)
for word in words:
    print(word, end=" ")


print(" ")    
sentence = "ALi:hassan:Sajad"
names = sentence.split(":")
print(names)
```

# Dictionary in For loop

```python
dict_1 = {1:"Apple", 2:"Cat", 3:"banana"}
for key in dict_1:
    print(key , end=" ")
print(" ")

for key in dict_1:
    print(f"{key} : {dict_1[key]}")
    
for value in dict_1.values():
    print(value , end=" ")
```
