```python
import matplotlib.pyplot as plt
```

```python
# Example_1:
import numpy as np

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Example_2:
mac_price = [200,250,300,350,330.5,380,400,399]
mic_price = [400,450,500,520,580,600,630,680]
years = [2015,2016,2017,2018,2019,2020,2021,2022]
```

```python
plt.plot(years, mac_price)
plt.scatter(years, mic_price)
# plt.plot(years, mac_price,':g'
#         ,years, mic_price,'--k')
plt.xlabel("Years")
plt.ylabel("Price")
plt.axis([2010,2025,100,1000])
plt.show()
```

```python
#Example_3:
fig_1 = plt.figure(1, figsize=(10,3))

chart_1=fig_1.add_subplot(223)
chart_2=fig_1.add_subplot(224)

chart_1.plot(years,mic_price)

chart_2.plot(years,mac_price)

plt.show()
```

# Subplot

```python
# Example_4:
plt.subplot(121)
plt.plot(years, mac_price)
plt.title("Price of mac")
plt.xlabel("years")
plt.ylabel("price")

plt.subplot(122)
plt.plot(years, mic_price)
plt.title("price of mic")

plt.suptitle("PRICE")

plt.show()
```

# Chart Title 

```python
x = np.array([80,85,90,95,100,105])
y = np.array([200,220,300,330,450,160])

font_1 = {'family':'serif','color':'blue','size':'20'}
font_2 = {'family':'serif','color':'darkred','size':'15'}

plt.plot(x,y)

# plt.title("Chart Example", fontdict=font_1, loc='left')
plt.title("Chart Example", fontdict=font_1)
plt.xlabel("X label", fontdict=font_2)
plt.ylabel("Y label", fontdict=font_2)

plt.show()
```

# Grid

```python
x = np.array([20,30,40,50,60,70,80])
y = np.array([240,250,260,270,280,290,300])
font_1 = {'color':'blue'}

plt.title("Chart Example")
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.plot(x,y)
plt.grid(color='green', linestyle='--',linewidth = "1")
# plt.grid(axis='x')
# plt.grid(axis='y')

plt.show()
```

# Scatter

```python
colors = np.array(["red","green","blue", "yellow","brown","purple","gray"])

x = np.array([20,30,40,50,60,70,80])
y = np.array([240,250,260,270,280,290,300])
# plt.scatter(x,y,color="#9c6292")
plt.scatter(x,y,c=colors)

xx = np.array([13,15,18,33,44])
yy = np.array([280,250,200,300,125])
plt.scatter(xx,yy,color="#629c88")

plt.colorbar()

plt.show()

```

# Bar

```python
x = ["A","B","C","D"]
y = [3,10,15,2]
plt.bar(x,y,color="red",width = 0.5)
plt.show()
```
